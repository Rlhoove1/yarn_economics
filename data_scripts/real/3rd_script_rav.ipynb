```python
import requests
import json
import time
import pandas as pd
```

```python
url = f"https://api.ravelry.com/projects/search.json"
uid, pw = "read-0ea4f9f35434b031dbcf2acbe301c226", "tlSskeRXLNybZBFYzE2gZaAL7K2Fh/eXPnLuHly9"
```

```python
#https://www.ravelry.com/projects/search#craft=crochet&happiness=4&photo=yes&status=finished&meterage-in=-1500-1800%2B-1800-2100%2B-2100-%2B-0-150%2B-150-300%2B-300-450%2B-450-600%2B-600-750%2B-750-900%2B-900-1200%2B-1200-1500&hooks=5.0mm&finished-in=1&country=united-states&sort=completed&view=thumbs
# 93,147 matches
#not have meterage, and has 5.0mm
#6
df = pd.DataFrame()

# Prepqre
response = requests.get(url, auth=(uid, pw), params={
    "page": 1,
    "page_size": 4999,
    "craft": "crochet",
    "photo": "yes",
    "country": "united-states",
    "happiness": "4",
    "finished-in": "1",
    "status": "finished",
    "meterage-in": "-1500-1800+-1800-2100+-2100-+-0-150+-150-300+-300-450+-450-600+-600-750+-750-900+-900-1200+-1200-1500",
    "hooks": "5.0mm"
})
last_page = response.json()['paginator']['last_page']
print(f"Total pages: {last_page}")

for page in range(1, last_page + 1):
    print(f"Fetching page {page}/{last_page}...")

    retry_delay = 60
    while True:
        response = requests.get(url, auth=(uid, pw), params={
            "page": page,
            "page_size": 4999,
            "craft": "crochet",
            "photo": "yes",
            "country": "united-states",
            "happiness": "4",
            "finished-in": "1",
            "status": "finished",
            "meterage-in": "-1500-1800+-1800-2100+-2100-+-0-150+-150-300+-300-450+-450-600+-600-750+-750-900+-900-1200+-1200-1500",
            "hooks": "5.0mm"
        })

        try:
            responselist = response.json()
            break
        except requests.exceptions.JSONDecodeError:
            print(f"  Empty response (status: {response.status_code}), retrying in {retry_delay}s...")
            time.sleep(retry_delay)
            retry_delay = min(retry_delay * 2, 300)

    df_page = pd.json_normalize(responselist['projects'])
    df = pd.concat([df, df_page], ignore_index=True)
    print(f"  → {len(df_page)} records (total: {len(df)})")

    if page < last_page:
        time.sleep(60)

print(f"Done. Total rows: {len(df)}")
```
