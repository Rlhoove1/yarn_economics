```python
import requests
import json
import time
import pandas as pd
```

```python
url = f"https://api.ravelry.com/projects/search.json"
uid, pw = "read-0ea4f9f35434b031dbcf2acbe301c226", "tlSskeRXLNybZBFYzE2gZaAL7K2Fh/eXPnLuHly9"
```

```python
#https://www.ravelry.com/projects/search#craft=crochet&happiness=4&photo=yes&status=finished&finished-in=1&country=united-states&meterage-in=0-150&sort=completed&view=thumbs
# 88,810 matches
df_1 = pd.DataFrame()

# Prepqre
response = requests.get(url, auth=(uid, pw), params={
    "page": 1,
    "page_size": 4999,
    "craft": "crochet",
    "photo": "yes",
    "country": "united-states",
    "happiness": "4",
    "finished-in": "1",
    "status": "finished",
    "meterage-in": "0-150"
})
last_page = response.json()['paginator']['last_page']
print(f"Total pages: {last_page}")

for page in range(1, last_page + 1):
    print(f"Fetching page {page}/{last_page}...")

    retry_delay = 60
    while True:
        response = requests.get(url, auth=(uid, pw), params={
            "page": page,
            "page_size": 4999,
            "craft": "crochet",
            "photo": "yes",
            "country": "united-states",
            "happiness": "4",
            "finished-in": "1",
            "status": "finished",
            "meterage-in": "0-150"
        })

        try:
            responselist = response.json()
            break
        except requests.exceptions.JSONDecodeError:
            print(f"  Empty response (status: {response.status_code}), retrying in {retry_delay}s...")
            time.sleep(retry_delay)
            retry_delay = min(retry_delay * 2, 300)

    df_page = pd.json_normalize(responselist['projects'])
    df_1 = pd.concat([df_1, df_page], ignore_index=True)
    print(f"  → {len(df_page)} records (total: {len(df_1)})")

    if page < last_page:
        time.sleep(60)

print(f"Done. Total rows: {len(df_1)}")
```

```python
df_1.to_csv("progress_save_part2_88670.csv")
```
